**Handling Null valus**

```python
filter_df["description"].fillna("Not mentioned",inplace=True)
filter_df.dropna(how="any", subset=["name"],inplace=True)
filter_df["neighborhood_overview"].fillna("Not mentioned",inplace=True)
filter_df["house_rules"].fillna("Not mentioned",inplace=True)
filter_df["weekly_price"].fillna(0,inplace = True)
filter_df["monthly_price"].fillna(0,inplace = True)
filter_df["cleaning_fee"].fillna(0,inplace=True)
filter_df["last_review"].fillna("Not mentioned",inplace=True)
filter_df["review_scores.review_scores_cleanliness"].fillna(0,inplace=True)
filter_df["review_scores.review_scores_communication"].fillna(0,inplace=True)
filter_df["minimum_nights"].fillna(0,inplace=True)
filter_df["security_deposit"].fillna(0,inplace=True)

filter_df.isna().sum()
```
